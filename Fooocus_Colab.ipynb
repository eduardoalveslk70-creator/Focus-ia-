````python
# 🔵 OPCIONAL — Montar Google Drive para salvar modelos
# Execute esta célula ANTES da célula 6 se quiser persistir os modelos.
# Assim não precisa baixar ~7GB toda vez que o runtime reiniciar.

from google.colab import drive
drive.mount('/content/drive')

import os
models_drive = '/content/drive/MyDrive/Fooocus_models'
models_local = '/content/Fooocus/models'

os.makedirs(models_drive, exist_ok=True)

# Cria symlink para salvar/carregar modelos direto do Drive
if not os.path.islink(models_local):
    if os.path.exists(models_local):
        import shutil
        shutil.move(models_local, models_drive)
    os.symlink(models_drive, models_local)
    print('✅ Modelos vinculados ao Google Drive!')
else:
    print('✅ Symlink já existe, modelos sendo carregados do Drive.')
````
